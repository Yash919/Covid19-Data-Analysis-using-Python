**Import the Packages**

```python
import pandas as pd
import numpy as np
import seaborn as sns   #Seaborn is an open-source Python library built on top of matplotlib. It is used for data visualization and exploratory data analysis.
                        # It will be used to visualize random distributions.
import matplotlib.pyplot as plt
print("Modules are imported")
```

1.1 **Importing Covid-19 Dataset**

```python
corona_dataset_csv=pd.read_csv("/content/Covid19-Data-Analysis-using-Python/covid19_Confirmed_dataset.csv")
corona_dataset_csv.head(10)
```

```python
corona_dataset_csv.shape
```

```python
columns=corona_dataset_csv.columns
columns
```

1.2 **Dropping the unnecessary columns**

```python
corona_dataset_csv.drop(['Lat','Long'],
                        axis=1,
                        #default value, annotation axis=0 which is equal to rows
                        inplace=True #will change the corona dataset too
                        )
```

```python
corona_dataset_csv.head(10)
```

1.3 **Aggregating the rows by the country**

```python
corona_dataset_aggregate=corona_dataset_csv.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregate.head()
#After aggregation, the index of the df is the column at which we aggregated
```

```python
corona_dataset_aggregate.shape
```

1.4 **Visualizing data related to a country**

```python
corona_dataset_aggregate.loc['China']
```

1.5 **Plotting china,egypt,spain,India aggregate sum**

```python
corona_dataset_aggregate.loc['China'].plot()
corona_dataset_aggregate.loc['Egypt'].plot()
corona_dataset_aggregate.loc['Italy'].plot()
corona_dataset_aggregate.loc['India'].plot()
plt.legend()
```

2.1 **Calculating the first derivative of the curve**

```python
corona_dataset_aggregate.loc["China"].diff().plot()
```

2.2 **Finding maximum infection rate [China,Italy,Spain,India]**

```python
print("The maximum infection rate in China:",corona_dataset_aggregate.loc["China"].diff().max())
print("The maximum infection rate in Italy:",corona_dataset_aggregate.loc["Italy"].diff().max())
print("The maximum infection rate in Spain:",corona_dataset_aggregate.loc["Spain"].diff().max())
print("The maximum infection rate in India:",corona_dataset_aggregate.loc["India"].diff().max())
```

2.3 **Finding all the Countries max infection rate**

Bonus Question - **Find which country have max infection rate**

```python
countries=list(corona_dataset_aggregate.index)
maxi=0
country_name=None
max_infection_rate=[]
for country in countries:
  max_infection_rate.append(corona_dataset_aggregate.loc[country].diff().max())
  if(maxi<max_infection_rate[-1]):
    maxi=max_infection_rate[-1]
    country_name=country
print(countries)
print(max_infection_rate)
print(country_name,maxi)
```

2.4 **Adding Column in a dataset (max_infection_rate)**

```python
corona_dataset_aggregate["max_infection_rate"]=max_infection_rate
corona_dataset_aggregate.head()
```

3 **Making New Dataframe and Importing new Dataset [Worldwide_happiness_report]**

```python
corona_data=pd.DataFrame(corona_dataset_aggregate["max_infection_rate"])
corona_data.head()
```

```python
happiness_report_csv=pd.read_csv("/content/Covid19-Data-Analysis-using-Python/worldwide_happiness_report.csv")
happiness_report_csv.head()
```

3.1 **Dropping Columns**

[a column with very low variance can be dropped from the data because such column does not help in predicting the target variable.]

```python
# No need of corruption,generosity,overall_rank,score
remove_columns=["Perceptions of corruption","Score","Overall rank","Generosity"]
happiness_report_csv.drop(remove_columns,axis=1,inplace=True)
happiness_report_csv.head()
```

```python
happiness_report_csv.set_index("Country or region",inplace=True)
happiness_report_csv.head()
```

3.2 **Joining two datasets [corona_data,happiness_report_csv]**

```python
#Rows and Columns in a dataset.
print(happiness_report_csv.shape)
print(corona_data.shape)
```

```python
#Applying inner-join common-attribute in both [Country or region]
data=corona_data.join(happiness_report_csv,how="inner")
data.head()
```

3.3 **Correlation matrix**

```python
data.corr()        #As all values are +ve we can plot graph any of using attribute
```
